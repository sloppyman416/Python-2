## Handling Missing data

````python
import pandas as pd
import numpy as np
````

````python
data={
    "Name":['A','B','C','D','E'],
    "Marks":[85,np.nan,78,np.nan,90],
    "Bonus":[5,7,np.nan,6,5],
    "City":['Ahmedabad','Surat','Rajkot','Ahmedabad',None]
}

df=pd.DataFrame(data)
df
````

## fillna() Method

````python
df.fillna(100)
````

````python
df
````

## Check Missing Value

````python
df.isnull()
````

````python
df.isnull().sum()
````

## `dropna()`
- This is used to drop the rows/columns with missing value

**syntax**
```python
df.dropna(axis=0,how=,thresh=None,subset=[],inplace=True)
```
- axis=0 means rows and axis=1 means columns

````python
df.dropna(axis=0)
````

````python
df.dropna(axis=1)
````

### `how=`
- how='any'
    - removes rows with atleast one missing value
- how='all'
    - removes rows with all missing value

````python
df.dropna(how='any')
````

````python
df.loc[5]=[None,None,None,None]
df
````

````python
df.dropna(how='all')
````

### `thresh=`
- Minimum number of **non null values** to keep in a row

````python
df
````

````python
df.dropna(thresh=3)
````

````python
df.loc[0,'Marks']=np.nan
df
````

````python
df.loc[0,'Bonus']=np.nan
df
````

````python
df.dropna(thresh=3)
````

### `subset=`
- check missing value only in specific column

````python
df.dropna(subset=['Marks'])
````

````python
df.dropna(subset=['Marks','Bonus'])
````

### `drop()`
- it removes the specified row or column

````python
df
````

````python
df.drop(columns=['City','Marks'])
````

````python
df.drop(index=1)
````
