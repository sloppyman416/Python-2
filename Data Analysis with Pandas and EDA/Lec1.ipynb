## Pandas Library
- panal data system
- used for data manipulation

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````


## Create Series
- 1D pandas datastructure

````python
# using list
a=pd.Series([1,2,3,4,9,10])
print(a)
print(type(a))
````

````python
# using tuple
a=pd.Series((1,2,3,4,9,10))
print(a)
print(type(a))
````

````python
# using dictioary
a=pd.Series({
    'a':'Python',
    'b':'java'
})
print(a)
print(type(a))
````

````python
# using range
a=pd.Series(range(1,6))
print(a)
print(type(a))
````

## DataFrame
- A 2D pandas data structure
- includes rows and columns
- a group of multiple dataframes is called panel
- a group of multiple series is called dataframe

````python
a=pd.DataFrame({
    'name':['A','B','C'],
    'marks':[80,90,100]
})
print(a)
print(type(a))
````

````python
# using list of lists

a=pd.DataFrame([['A',80],['B',90],['C',100]],columns=['name','marks'])
print(a)
print(type(a))
````

````python
# using list of dictionaries
pd.DataFrame([
    {'name':'A','marks':80},
    {'name':'B','marks':90},
    {'name':'C','marks':100}
])
# print(a)
# print(type(a))
````

## Indexing and Slicing (Accessing elements)

````python
data={
    'name':['A','B','C','D','E'],
    'marks':[85,90,95,np.nan,99],
    'bonus':[1,2,3,4,7],
    'city':['ahm','raj','ahm','ahm','raj']
}
df=pd.DataFrame(data)
df
````
