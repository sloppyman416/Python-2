## Seaborn

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

## Heatmap

```python
df=pd.DataFrame({
    "Student":['A','B','C','D','E','F','G','H'],
    'MATH':[78,85,90,66,88,120,45,95],
    'SCIENCE':[80,82,89,72,85,110,50,92],
    'ENGLISH':[90,91,95,72,89,130,55,96],
    'COMPUTER':[75,88,92,60,84,115,48,90],
    'GENDER':['M','F','M','F','M','F','M','F'],
    'STUDYTYPE':['self',"coaching","self","self","coaching","self","coaching","self"],
    'HOURS':[3,5,4,1,2,7,8,6]
})
```

```python
df
```

### Create a Correlation Table

```python
corr=df[['MATH','SCIENCE','ENGLISH','COMPUTER','HOURS']].corr()
```

```python
corr
```

```python
sns.heatmap(
corr
)
```

```python
sns.heatmap(
corr,
vmin=0.3,
vmax=1.0,
cmap='viridis',
linewidths=2,
linecolor='green',
annot=True,
)
```

```python
sns.heatmap(
corr,
vmin=0.3,
vmax=1.0,
cmap='viridis',
linewidths=2,
linecolor='green',
annot=True,
cbar=False
)
plt.title("Seaaborn heatMap of correlation table")
plt.show()
```

### Seaborn based boxplot
- Boxplot is used to visualize statistical metrics like meann,median,max percentiles and outliers.

```python
sns.boxplot(
x='STUDYTYPE',
y='MATH',
data=df
)
plt.title('Seaborn based boxplot')
plt.show()
```

```python
sns.boxplot(
x='STUDYTYPE',
y='MATH',
data=df,
hue='GENDER'
)
plt.title('Seaborn based boxplot')
plt.show()
```

### Seaborn based scatter plot
- TO Visualize relationship between two numeric variables

```python
sns.scatterplot(
    x='HOURS',
    y='MATH',
    data=df,
    hue='GENDER',
    style='STUDYTYPE',
    s=100,
)
```
