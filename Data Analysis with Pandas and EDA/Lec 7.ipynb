## NETWORKX
- it is a library used in Graph Theory
- it is built on matplotlib

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Create Graph Object
G=nx.Graph()
```

```python
# add node
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(4,7))
# add edges
G.add_edge(1,2)
G.add_edges_from([(2,3),(3,6),(5,6),(4,6),(4,4)])
nx.draw_networkx(G,with_labels=True,node_size=1000,node_color="maroon",edge_color="green")
plt.show()
```

```python
g=nx.DiGraph()
# add nodes
g.add_nodes_from(["A","B","C","D"])
# add edges
g.add_edges_from([("A","B"),("B","C"),("B","D"),("C","D"),("D","A")])
nx.draw_networkx(g,with_labels=True)
plt.show()
```

```python
g=nx.DiGraph()
# add nodes
g.add_nodes_from(range(0,6))
# add edges
g.add_edges_from([(1,1),(1,5),(1,3),(1,2),(2,5),(2,4),(2,1),(2,0),(0,1),(0,2),(0,3)])
nx.draw_networkx(g,with_labels=True,node_color="pink")
plt.show()
```

## Graph based on adgecency matrix

```python
g=nx.DiGraph()
L=[[1,0,0,1,0,1,1],
   [0,1,1,1,1,1,0],
   [1,1,1,0,1,0,1]]
g.add_nodes_from([0,1,2])
for i in range(0,3):
    for j in range(0,3):
        if L[i][j]==1:
            g.add_edge(i,j)
nx.draw_networkx(g,with_labels=True)
plt.show()
```
