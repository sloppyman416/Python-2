## JSON
- JSON (JavaScript Object Notation) is a lightweight data-interchange format.
- for data transfer using serialization and deserialization
- JSON data is a collection of key/value pairs; in Python, it is realized by a dictionary.

```python
import json
import pprint
resp = req.request(method='GET', url="http://ip-api.com/json/")
print(resp.text)
print()
json_page = json.loads(resp.text)
print(json_page)
print()
pprint.pprint(json_page)
```

```python
file_name = "Sample.json"
with open(file_name, "w") as fid: 
     json.dump(json_page, fid)
```

### `beautifulsoup`
- the library is use to scrap static webpage

## extract webpage (online)

```python
from bs4 import BeautifulSoup
import requests
url="http://www.google.com"
response=requests.get(url)
content=response.text
soup=BeautifulSoup(content,'html.parser')
soup.prettify()
```

## extract webpage (offline - to be asked in exam)

```python
with open("BooksToScrape.html",encoding='utf-8') as file:
    html=file.read()
soup=BeautifulSoup(html,'html.parser')
print(soup.prettify())
```

```python
print("😀😀😀😀😀😀")
```

## access soup attributes

```python
print(soup.title)
print(soup.title.text)
print(soup.title.name)
```

## fetch the first book
- find method returns the first instance of the class object

```python
first_book=soup.find("article")
```

```python
print(first_book)
```

```python
first_price=soup.find(class_="price_color")
print(first_price)
print(first_price.text)
```

## Attributes

```python
print(first_book.attrs)
print(first_price.attrs)
```

## find all books
- `find_all` method returns all the instance of the tag for the given class

```python
all_books=soup.find_all('article')
print(all_books)   # in list form
print(len(all_books))
```
