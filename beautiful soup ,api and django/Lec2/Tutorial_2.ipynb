## Scraping a job portal - a hands on tutorial

```python
# import libraries
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
# read the page
url="https://realpython.github.io/fake-jobs/"
response=requests.get("https://realpython.github.io/fake-jobs/")
print(response)
```

```python
# read content
content=response.text
soup=BeautifulSoup(content,'html.parser')
soup
```

## access title

```python
print(soup.title)
print(soup.title.text)
print(soup.title.name)
```

## `find_all` method
- this method returns the list of all tags with given class name

```python
profiles=soup.find_all("h2")
for i in profiles:
    if "Python" in i.text:
        print(i.text)
    
```

```python
pythonjobs=soup.find_all("h2",string=lambda text:"python" in text.lower())
print(pythonjobs)
print()
for i in pythonjobs:
    print(i.text)
```

## `find` method
- this method returns the first instance of the given class and tag

```python
results=soup.find(id="ResultsContainer")
print(results)
```

## `get_text` method
- this method is used to get the text content of the webpage

```python
print(results.get_text())
```

```python
print(results.get_text(strip=True))
```

## access job details

```python
jobelements=soup.find_all("div",class_="card-content")
print(jobelements)
```

## Access the title, company and location of the jobs

```python
for i in jobelements:
    titletag=i.find("h2",class_="title")
    companytag=i.find("h3",class_="company")
    locationtag=i.find("p",class_="location")
    print(titletag.text.strip())
    print(companytag.text.strip())
    print(locationtag.text.strip())
```

## Fetch links  - `a` tags

```python
for i in jobelements:
    links=i.find_all("a")
    print(links[1])
    
```

```python
# fetch the link address
for i in jobelements:
    links=i.find_all("a")
    for link in links:
        linkurl=link['href']
        print(linkurl)
```

## extract table data

```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# read file
with open("worldpopulation.html",encoding="utf-8") as file:
    html=file.read()
soup=BeautifulSoup(html,"html.parser")    
soup
```

```python
table=soup.find('table',class_="wikitable")
print(table)
```

## find all the rows

```python
rows=table.find_all("tr")
print(rows)
```

## iterate over the rows

```python
for row in rows:
    cells=row.find_all("td")
    for cell in cells:
        print(cell.text.strip(),end='\t')
    print()
```

## extract topnews from https://indianexpress.com/

```python
# read the page
url="https://indianexpress.com/"
response=requests.get(url)
print(response)
```

```python
# read content
content=response.text
soup=BeautifulSoup(content,'html.parser')
soup
```

```python
# extract top news
topnews=soup.find("div",id="HP_TOPNEWS")
topnews
```

```python
print(topnews.get_text())
```

## find the links

```python
links=topnews.find_all("a")
links
```

```python
for i in links:
    headline=i.text.strip()
    print(headline)
```
