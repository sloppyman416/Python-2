#### SVM (SUPPORT VECTOR MACHINE) (CONCEPT OF HYPERPLANE)

- a class of machine learning algorithm use in classification and regression
- it uses the concept of hyperplane / decision boundary

#### python implementation

```python
import numpy as np
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
# data
x=np.array([[1,6],[2,7],[6,3],[7,2]])
y=np.array([-1,-1,1,1])
```

```python
# model
model=SVC(kernel='linear')
```

```python
# fit
model.fit(x,y)
```

```python
# parameters
w=model.coef_[0]
b=model.intercept_[0]
print("slope/weight = ",w)
print("intercept/bias = ",b)
```

#### plot the scatter and decision boundary

```python
plt.scatter(x[:,0],x[:,1],c=y)
# decision boundary
xval=np.linspace(0,8,100)
yval=-(w[0]*xval+b)/w[1]
plt.plot(xval,yval)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("SVM with decision boundary")
plt.show()
```

```python
# prediction
newpoints=np.array([[3,6],[5,3],[4,5]])
model.predict(newpoints)
```

### KNN algorithm
- this class of algorithms are used in classification problems of machine learning
- it is based on the concept of nearest neighbors (Eucleadian distance)

### algorithm steps
1. decide the number of k parameter (nearest neighbors)
2. calculate the distance between query point and all the given features
3. sort the distance and determine the rank based on minimum distance
4. gather category of nearest neighbour
5. use the voting to predict the category of query point

#### python implementation of knn algorithm

```python
# import libraries
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
# define features and target
x=np.array([[7,7],[7,4],[3,4],[1,4]])
y=np.array(['Bad','Bad','Good','Good'])
```

```python
# create model
knn=KNeighborsClassifier(n_neighbors=3)
```

```python
# fit the model
knn.fit(x,y)
```

```python
# predict for the query point
knn.predict([[3,7]])
```

Write Python code to train a kNN classifier using the following steps:
Split the dataset X into training and testing sets with a test size of 0.3 and a random state of 42.
Initialize a kNN classifier with 5 neighbors.
Train the classifier on the training set.
Make predictions on the test set.
Calculate and print the accuracy score of the classifier.

```python
# import libraries
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# assuming the dataset is loaded and x and y are defined
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
# create the model
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

 You are tasked with using the k-Nearest Neighbors (kNN) algorithm to classify whether patients have diabetes or not based on 
certain diagnostic measurements. You have been provided with diabetes.csv file. The datasets consist of several medical 
predictor (independent) variables and one target (dependent) variable, Outcome. Independent variables include the number of 
pregnancies the patient has had, their BMI, insulin level, age, and so on. Also perform Model Performance Analysis using

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/patelmanishv/Sem4Data/refs/heads/main/Datasets/diabetes.csv")
df.head(2)
```

```python
x=df.drop(columns=["Outcome"])
y=df["Outcome"]
print(x.shape)
print(y.shape)
```

```python
# assuming the dataset is loaded and x and y are defined
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
# create the model
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

```python
predictons=knn.predict([[1,200,180,25,0,30,0.342,45]])
print(predictons)
print(confusion_matrix(y_test,y_pred))
```
