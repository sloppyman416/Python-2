### Multiple Linear Regression
- in multiple lineaar regression there are multiple features or independent variables and one dependent variable (target)

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
data=pd.DataFrame({
    'size':[2104,1600,2400,1416,3000],
    'year':[2003,1977,2001,1995,2000],
    'price':[399900,329900,369000,232000,539900]
})
```

```python
# feature and target
x=data[['size','year']]
y=data['price']
print(x.shape)
print(y.shape)
```

```python
# create model
model=LinearRegression()

# fit the model
model.fit(x,y)

# SLR slope and intercept
print("slope=",model.coef_[0]) #slope
print("intercept=",model.intercept_) #intercept
#r2 score
prediction=model.predict(x)
print(r2_score(y,prediction))
print(mean_squared_error(y,prediction))
```

```python
new_input=pd.DataFrame({
    'size':[750],
    'year':[2009]
})
prediction=model.predict(new_input)
print("Prediction for 750 sq ft area ,2009 is",prediction[0])
```

### Polynomial Regression

```python
# import libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
# load the data set
iris=sns.load_dataset('iris')
print(iris.shape)
# define features and target
x=iris[['sepal_length']]
y=iris['petal_length']
print(x.shape)
print(y.shape)
# create polynomial features
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
model=LinearRegression()
model.fit(x_poly,y)
print(model.coef_)
print(model.intercept_)
```

```python
x_test_value=np.linspace(4,8,100).reshape(-1,1)
# print(x_test_value)
x_test=pd.DataFrame(x_test_value,columns=['sepal_length'])
x_test_poly=poly.transform(x_test)
y_pred=model.predict(x_test_poly)
print(y_pred)
```
